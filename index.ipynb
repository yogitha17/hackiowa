```python
import pandas as pd
import numpy as np

df = pd.read_csv("Merged Data.csv")

df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Extract just the date part for a new 'Just_Date' column
df['Just_Date'] = df['Date'].dt.date

# Group by 'Just_Date' and sum the GHI values for each group
grouped = df.groupby('Just_Date')['GHI'].mean().reset_index()

# Show the grouped DataFrame
print(grouped)

# Save the grouped DataFrame to a new CSV file
grouped.to_csv('grouped.csv', index=False)
```

```python
import pandas as pd

# Read both CSV files into DataFrames
df1 = pd.read_csv("Bus_daily_avg.csv")
df2 = pd.read_csv("grouped.csv")



# Convert the date columns to datetime objects
df1['Date'] = pd.to_datetime(df1['Date']).dt.date
df2['Just_Date'] = pd.to_datetime(df2['Just_Date']).dt.date

# Merge the DataFrames based on the date columns
merged_df = pd.merge(df1, df2, left_on='Date', right_on='Just_Date', how='inner')

# drop the 'Just_Date' column
merged_df.drop('Just_Date', axis=1, inplace=True)

# export to csv
merged_df.to_csv('merged.csv', index=False)
```

```python

df = pd.read_csv("merged.csv")

df['Date'] = pd.to_datetime(df['Date'])

# Extract month and year from Date
df['YearMonth'] = df['Date'].dt.to_period('M')

# Group by Year-Month and aggregate
grouped_df = df.groupby('YearMonth').agg({
    'daily_avg': ['avg'],
    'GHI': ['avg']
}).reset_index()

# The grouped DataFrame will contain mean and sum of daily_avg and GHI for each month
print(grouped_df)

# Save the grouped DataFrame to a new CSV file
grouped_df.to_csv('grouped_month.csv', index=False)
```

```python
# Extract Year from Date
df['Year'] = df['Date'].dt.to_period('Y')

# Group by Year and aggregate
grouped_df_year = df.groupby('Year').agg({
    'daily_avg': ['mean'],
    'GHI': ['mean']
}).reset_index()

# The grouped DataFrame will contain mean and sum of daily_avg and GHI for each year
print(grouped_df_year)

# Save the grouped DataFrame to a new CSV file
grouped_df_year.to_csv('grouped_year.csv', index=False)
```
